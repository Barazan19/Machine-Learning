# Logistic Regression Using scikit-learn
This notebook demonstrates how to fit a logistic regression model using `scikit-learn` on the insurance claim dataset.

## Steps:
1. Load and visualize the data
2. Train logistic regression
3. Make predictions
4. Evaluate model
5. Plot the sigmoid curve

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/Insurance_Claim_Dataset.csv")
X = df[['Age']]
y = df['Claim']

plt.scatter(X, y, alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Claim")
plt.title("Insurance Claim Data")
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X, y)
```

```python
y_proba = model.predict_proba(X)[:, 1]
y_pred = model.predict(X)
```

```python
theta_0 = model.intercept_[0]
theta_1 = model.coef_[0][0]

print(f"Theta 0 (Intercept): {theta_0:.4f}")
print(f"Theta 1 (Slope): {theta_1:.4f}")
```

```python
from sklearn.metrics import log_loss, confusion_matrix, accuracy_score

log_loss_value = log_loss(y, y_proba)
conf_matrix = confusion_matrix(y, y_pred)
accuracy = accuracy_score(y, y_pred)

print(f"Log Loss: {log_loss_value}")
print("Confusion Matrix:")
print(conf_matrix)
print(f"Accuracy: {accuracy}")
```

```python
import numpy as np

x_vals = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
y_vals = model.predict_proba(x_vals)[:, 1]

plt.figure(figsize=(8, 6))
plt.scatter(X, y, label="Data", alpha=0.5)
plt.plot(x_vals, y_vals, color="red", label="Sigmoid Fit (sklearn)")
plt.title("Logistic Regression Fit with scikit-learn")
plt.xlabel("Age")
plt.ylabel("Probability of Claim")
plt.legend()
plt.grid(True)
plt.show()
```
