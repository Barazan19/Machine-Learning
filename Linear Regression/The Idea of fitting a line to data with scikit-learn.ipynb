
## Comparison: Manual vs scikit-learn Linear Regression

To validate the implementation of linear regression using gradient descent, we compare it against `scikit-learn`'s `LinearRegression`. This helps verify if the manually calculated slope (w), intercept (b), and MSE are close to the library's results.


```python

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# Prepare the data
df = pd.read_csv("data/insurance.csv")
X = df[['age']].values
y = df['charges'].values

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Extract parameters
w_lib = model.coef_[0]
b_lib = model.intercept_

# Predict and evaluate
y_pred = model.predict(X)
mse_lib = mean_squared_error(y, y_pred)

print("Slope (w):", w_lib)
print("Intercept (b):", b_lib)
print("MSE:", mse_lib)
```

```python
import matplotlib.pyplot as plt

# Scatter plot data asli
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', alpha=0.5, label='Actual Data')

# Plot hasil prediksi
plt.plot(X, y_pred, color='red', linewidth=2, label='Fitted Line (Scikit-Learn)')

# Label dan judul
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Linear Regression with Scikit-Learn (Age vs Charges)")
plt.legend()
plt.grid(True)
plt.show()
```
